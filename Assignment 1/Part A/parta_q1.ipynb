```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import regularizers, optimizers
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
```

```python
NUM_CLASSES = 3

epochs = 1000
batch_size = 32
num_neurons = 10
seed = 10

np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# scale data
def scale(X, X_min, X_max):
    return (X - X_min)/(X_max-X_min)
```

```python
#read train data
train_input = np.genfromtxt('ctg_data_cleaned.csv', delimiter= ',')
X, y = train_input[1:, :21], train_input[1:,-1].astype(int)
X = scale(X, np.min(X, axis=0), np.max(X, axis=0))
y = y-1

# Split data into 70:30 ratio for training and testing
trainX, testX, trainY, testY = train_test_split(X, y, test_size=0.3, random_state=1)
print(trainX.shape, testX.shape, trainY.shape, testY.shape)
```

```python
# Do k-fold cross validation
n_folds = 5
n_fig = 1
kfold = KFold(n_splits=n_folds, random_state=1, shuffle=True)

# K-fold Cross Validation model evaluation
histories = {}
fold_no = 1
reg = regularizers.l2(10e-6)
opt = keras.optimizers.SGD(learning_rate=0.01)
models_list = []

for train, val in kfold.split(trainX, trainY):
    model = keras.Sequential([keras.layers.Dense(num_neurons, kernel_regularizer=reg, activation='relu'),
                              keras.layers.Dense(NUM_CLASSES, activation='softmax')])

    model.compile(optimizer=opt,
                  loss=keras.losses.SparseCategoricalCrossentropy(),
                  metrics=['accuracy'])
    # Generate a print
    print('------------------------------------------------------------------------')
    print(f'Training for Fold {fold_no} ...')

    histories[fold_no] = model.fit(trainX[train], trainY[train], epochs=epochs,
                                   validation_data=(trainX[val], trainY[val]),
                                   verbose=2, batch_size=batch_size)
    fig_num = [1, 2, 3, 4]
    fig_metric = ['accuracy', 'val_accuracy', 'loss', 'val_loss']

    # plot all 5 folds learning curves in one figure
    for num, metric in zip(fig_num, fig_metric):
        plt.figure(num)  # Training Accuracy
        plt.plot(histories[fold_no].history[metric], label='Fold ' + str(fold_no))
    fold_no = fold_no + 1

fig_num = [1, 2, 3 , 4]
fig_title = ['Training Accuracy', 'Validation Accuracy', 'Training Loss', 'Validation Loss']
file_name = ['train_acc', 'val_acc', 'train_loss', 'val_loss']
legend_pos = ["lower right", "lower right", "upper right", "upper right"]

for num, title, filename, legendpos in zip(fig_num, fig_title, file_name, legend_pos):
    plt.figure(num)
    plt.title(title + ' for all 5 folds')
    plt.ylabel(title)
    plt.xlabel('Number of Epochs')
    plt.legend(loc=legendpos)
    plt.savefig('./plots/1a_' + filename + '.png')
```

```python
print('Final Validation Accuracy for all 5 folds:')
val_acc = []
for i in range(1 ,6):
    val_acc.append(histories[i].history['val_accuracy'][epochs-1])
    print(f'Fold {i}: {val_acc[i-1]}')
print('Mean  : ' + str(np.mean(val_acc)))
    
print()
print('Final Training Accuracy for all 5 folds:')
train_acc = []
for i in range(1 ,6):
    train_acc.append(histories[i].history['accuracy'][epochs-1])
    print(f'Fold {i}: {train_acc[i-1]}')
print('Mean  : ' + str(np.mean(train_acc)))


print()
print('Final Validation Loss for all 5 folds:')
val_loss = []
for i in range(1 ,6):
    val_loss.append(histories[i].history['val_loss'][epochs-1])
    print(f'Fold {i}: {val_loss[i-1]}')
print('Mean  : ' + str(np.mean(val_loss)))

print()
print('Final Training Loss for all 5 folds:')
train_loss = []
for i in range(1 ,6):
    train_loss.append(histories[i].history['loss'][epochs-1])
    print(f'Fold {i}: {train_loss[i-1]}')
print('Mean  : ' + str(np.mean(train_loss)))
```

```python
# Train without 5-fold - use 70 training and 30 test
model_data = {}
reg = regularizers.l2(10e-6)
opt = keras.optimizers.SGD(learning_rate=0.01)

model = keras.Sequential([keras.layers.Dense(num_neurons, kernel_regularizer=reg, activation='relu'),
                          keras.layers.Dense(NUM_CLASSES, activation='softmax')])

model.compile(optimizer=opt,
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])


model_data['1a_model'] = model.fit(trainX, trainY, epochs=epochs,
                                  validation_data=(testX, testY),
                                   verbose=2, batch_size=batch_size)
```

```python
# plot learning curves
plt.figure(1)
plt.plot(model_data['1a_model'].history['accuracy'], label='training accuracy')
plt.plot(model_data['1a_model'].history['val_accuracy'], label='test accuracy')
plt.ylabel('Train accuracy')
plt.xlabel('Number of Epochs')
plt.legend(loc="lower right")
plt.savefig('./plots/parta/q1_v2/upload/1a_acc_no_cv.png')

plt.figure(2)
plt.plot(model_data['1a_model'].history['loss'], label='training loss')
plt.plot(model_data['1a_model'].history['val_loss'], label='test loss')
plt.ylabel('Loss')
plt.xlabel('Number of Epochs')
plt.legend(loc="upper right")
plt.savefig('./plots/parta/q1_v2/upload/1a_loss_no_cv.png')
```

```python
print('Final results from the last epoch')
acc = model_data['1a_model'].history['accuracy'][epochs-1]
val_acc = model_data['1a_model'].history['val_accuracy'][epochs-1]
loss = model_data['1a_model'].history['loss'][epochs-1]
val_loss = model_data['1a_model'].history['val_loss'][epochs-1]
    
print()
print(f'Accuracy\t\t: {acc}')
print(f'Validation Accuracy\t: {val_acc}')
print(f'Loss\t\t\t: {loss}')
print(f'Validation Loss\t\t: {val_loss}')
```
